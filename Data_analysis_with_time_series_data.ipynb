```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_path = "/content/opsd_germany_daily.csv"
opsd_daily = pd.read_csv(data_path, index_col=0, parse_dates=True)

opsd_daily.head()
```

```python
opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday Name'] = opsd_daily.index.day_name()

opsd_daily.head()
```

```python
sns.set(rc={'figure.figsize':(11,4)})
opsd_daily['Consumption'].plot(linewidth=0.5)
```

```python
cols_plot = ['Consumption', 'Solar', 'Wind']
axes = opsd_daily[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11,9), subplots=True)
print(axes)
for ax in axes:
  ax.set_ylabel('Daily Totals (Gwh)')
plt.show()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(11,10), sharex=True)
print(axes)
for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
  sns.boxplot(data=opsd_daily, x="Month", y=name, ax=ax)
  ax.set_ylabel('GWh')
  ax.set_title(name)
  if ax != axes[-1]:
    ax.set_xlabel('')
plt.show()
```

```python
times_sample = pd.to_datetime(['2013-02-03', '2013-02-06', '2013-02-08'])
cons = opsd_daily.loc[times_sample, ['Consumption']].copy()
cons
```

```python
cons_freq = cons.asfreq('D')
cons_freq['Consumption freq'] = cons.asfreq('D', method='ffill')
cons_freq
```

```python
data_cols = ["Consumption", "Wind", "Solar", "Wind+Solar"]
opsd_weekly_mean = opsd_daily[data_cols].resample('W').mean()
opsd_weekly_mean.head(10)
```

```python
print(opsd_daily.shape[0])
print(opsd_weekly_mean.shape[0])
```

```python
start, end = '2017-01', '2017-06'

fig, ax = plt.subplots()
ax.plot(opsd_daily.loc[start:end, 'Solar'], marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'], marker='o', markersize=8, linestyle='-', label='Weekly mean sample')
ax.set_ylabel('Solar Production (GWh)')
ax.legend()
plt.show()
```

```python
opsd_annual = opsd_daily[data_cols].resample('Y').sum(min_count=360)
opsd_annual = opsd_annual.set_index(opsd_annual.index.year)
opsd_annual.index.name = "Year"
opsd_annual["Wind+Solar/Consumption"] = opsd_annual["Wind+Solar"]/opsd_annual["Consumption"]
opsd_annual
```

```python
ax = opsd_annual.loc[2012:, "Wind+Solar/Consumption"].plot.bar(color="C0")
ax.set_ylabel("Fraction")
ax.set_ylim(0, 0.3)
ax.set_title("Wind + Solar Share of Annual Electricity Consumption")
plt.xticks(rotation=0)
plt.show()
```

```python
opsd_7day = opsd_daily[data_cols].rolling(7, center=True).mean()
opsd_7day.head(10)
```

```python
opsd_365day = opsd_daily[data_cols].rolling(window=365, center=True, min_periods=360).mean()
opsd_365day.head(10)
```

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots()
ax.plot(opsd_daily["Consumption"], marker='.', markersize=2, color='0.6', linestyle='None', label="Daily")
ax.plot(opsd_7day["Consumption"], linewidth=2, label="7-day Rolling Mean")
ax.plot(opsd_365day["Consumption"], linewidth=3, color="0.3", label="365-day Rolling Mean")
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel("Year")
ax.set_ylabel("Consumption")
ax.set_title("Trends in Electricity Consumption")
plt.show()
```

```python
fig, ax = plt.subplots()
for mn in ['Wind', 'Solar', 'Wind+Solar']:
  ax.plot(opsd_365day[mn], label=mn)
  ax.xaxis.set_major_locator(mdates.YearLocator())
  ax.legend()
  ax.set_ylabel("Production")
  ax.set_title("Trends in Electricity Production (365-d Rolling Means)")
plt.show()
```
